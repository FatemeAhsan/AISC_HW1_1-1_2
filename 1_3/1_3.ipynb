```python
import numpy as np
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from sklearn.linear_model import LinearRegression
```

```python
ws = load_workbook('linear_regression_data.xlsx', data_only=True)['Sheet1']
data = []
for i in range(2,  ws.max_row + 1):
    data.append([ws[f'A{i}'].value, ws[f'B{i}'].value,  ws[f'C{i}'].value,  ws[f'D{i}'].value])
data = np.array(data)
```

```python
data
```

```python
x = data[:, :3]
y = data[:, 3].reshape(-1, 1)
```

```python
x, y
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.coef_, model.intercept_
```

```python
plt.plot(x[:, 0], y, 'o', color='r')
plt.plot(x[:, 1], y, 'o', color='g')
plt.plot(x[:, 2], y, 'o', color='b')
```

```python
x_smp = [750, 750, 750]
x_smp = np.array([x_smp])
y_smp = model.predict(x_smp)
y_smp
```

```python
y_prd = model.predict(x)
```

```python
y, y_prd
```

```python
plt.plot(x[:, 0], y, 'o', label='True')
plt.plot(x[:, 0], y_prd, 'o', label='Predicted')
plt.legend()
```

```python
plt.plot(x[:, 1], y, 'o', label='True')
plt.plot(x[:, 1], y_prd, 'o', label='Predicted')
plt.legend()
```

```python
plt.plot(x[:, 2], y, 'o', label='True')
plt.plot(x[:, 2], y_prd, 'o', label='Predicted')
plt.legend()
```
